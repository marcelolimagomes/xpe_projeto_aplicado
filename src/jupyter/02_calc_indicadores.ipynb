```python
# Recupera base de dados com histórico de cotações de Ações e FII's baixadas previamente 
data_file_list = utils.get_data_files()

# Itera a lista de base de dados com histórico de cotações de Ações e FII's baixadas previamente 
for data_file in data_file_list:
	data = pd.read_csv(data_file, sep=';') # Lê base de histórico de cotações do ticket correspondente
	if data.shape[0] > 0:
		# Calcula o indicador EMA (Exponential Moving Average na base 200 periodos) para variável 'adj_close'
		data = utils.calc_ema_periods(data, periods_of_time=[200], close_price='adj_close', diff_price=True)
		# Calcula o indicador EMA (Exponential Moving Average na base 200 periodos) para variável 'close'
		data = utils.calc_ema_periods(data, periods_of_time=[200], close_price='close', diff_price=True)
		# Calcula o indicador RSI (Relative Strength Index) para variável 'adj_close'
		data = utils.calc_RSI(data, close_price='adj_close')
		# Arredonda valores para duas casas decimais
		data = data.round(2)
		# Grava base de dados com indicadores calculados
		data.to_csv(data_file, sep=';', index=False)
		print(f'Ticker: {data.tail(1).ticker.values[0]} - Shape {data.shape}')
		# data.info()

```
